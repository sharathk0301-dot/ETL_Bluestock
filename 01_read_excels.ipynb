```python
import pandas as pd
from pathlib import Path
import re

RAW_DIR = Path("data/raw")

companies_df     = pd.read_excel(RAW_DIR / "companies.xlsx",     header=1)
balancesheet_df  = pd.read_excel(RAW_DIR / "balancesheet.xlsx",  header=1)
profitandloss_df = pd.read_excel(RAW_DIR / "profitandloss.xlsx", header=1)
cashflow_df      = pd.read_excel(RAW_DIR / "cashflow.xlsx",      header=1)
analysis_df      = pd.read_excel(RAW_DIR / "analysis.xlsx",      header=1)
prosandcons_df   = pd.read_excel(RAW_DIR / "prosandcons.xlsx",   header=1)
documents_df     = pd.read_excel(RAW_DIR / "documents.xlsx",     header=1)
```

```python
companies_clean = companies_df.copy()

companies_clean.columns = (
    companies_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

dim_company = companies_clean[[
    "id",
    "company_name",
    "about_company",
    "website",
    "nse_profile",
    "bse_profile",
    "face_value",
    "book_value",
    "roce_percentage",
    "roe_percentage",
]].copy()

dim_company = dim_company.rename(columns={"id": "company_id"})
dim_company["company_id"] = dim_company["company_id"].astype(str).str.strip()

dim_company.info()
```

```python
profitandloss_clean = profitandloss_df.copy()

profitandloss_clean.columns = (
    profitandloss_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

profitandloss_clean["company_id"] = profitandloss_clean["company_id"].astype(str).str.strip()
profitandloss_clean["year"] = profitandloss_clean["year"].astype(str).str.strip()

profitandloss_clean["fiscal_year"] = (
    profitandloss_clean["year"]
    .str.extract(r"(\d{4})", expand=False)
)
profitandloss_clean["fiscal_year"] = profitandloss_clean["fiscal_year"].astype("float")

fact_profit_loss = profitandloss_clean[[
    "company_id",
    "fiscal_year",
    "sales",
    "expenses",
    "operating_profit",
    "opm_percentage",
    "other_income",
    "interest",
    "depreciation",
    "profit_before_tax",
    "tax_percentage",
    "net_profit",
    "eps",
    "dividend_payout",
]].copy()

fact_profit_loss["net_margin"] = (
    fact_profit_loss["net_profit"] / fact_profit_loss["sales"]
)

fact_profit_loss["operating_margin"] = (
    fact_profit_loss["operating_profit"] / fact_profit_loss["sales"]
)

fact_profit_loss.info()
```
